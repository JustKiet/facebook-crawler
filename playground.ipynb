```python
from smart_open import open
from bs4 import BeautifulSoup

with open('posts_html.html', 'r', encoding='utf-8') as f:
    text = f.read()

soup = BeautifulSoup(text, 'html.parser')

# Remove all rubbish divs
for div in soup.select('div[data-0="0"]'):
    div.decompose()

# Remove all rubbish spans (spans with content less than 2 characters)
for span in soup.find_all('span'):
    if len(span.text) < 2:
        span.decompose() 

# Specify tags to keep
target_tags = ["div", "h3", "h4", "span", "a"]

# Delete all other tags
for tag in soup.find_all():
    if tag.name not in target_tags:
        tag.unwrap()

"""for div in soup.find_all('div'):
    if len(div.attrs) == 0:
        div.unwrap()"""

for span in soup.find_all('span'):
    if len(span.attrs) == 0:
        span.unwrap()

with open('test.txt', 'w', encoding='utf-8') as f:
    f.write(soup.prettify())

posts = []

# Extract each post by finding all divs with only one class with its first child div having no attributes
for tag in soup.find_all('div'):
    if 'class' in tag.attrs and len(tag['class']) == 1:
        for child_div in tag.find_all('div'):
            if not child_div.attrs:
                if len(child_div.text) > 0:
                    tag_text = tag.text
                    posts.append(tag_text)
                    with open('posts_content.txt', 'a', encoding='utf-8') as f:
                        f.write(tag.text + "\n\n\n")
            break
posts[1]
```
